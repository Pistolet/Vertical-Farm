[**식물공장의 개념 및 시장현황**](#)
============================
 
**식물공장(plant factory)**이란, 농작물에 대하여 **"통제된 일정한 시설 내에서 빛, 온‧습도, 이산화탄소 농도 및 배양액 등의 환경 조건을 인공적으로 제어하여 계절이나 장소에 관계없이 자동적으로 연속 생산하는 시스템”**을 말한다.

식물공장(plant factory)이란 용어의 기원은 정확하지 않으나, 1960년대에 유럽에서 채소의 시설 재배가 성행하면서 공장식 농업을 식물공장이라고 불렀던 것으로 알려진다.  북유럽 지역은 흐린 날이 많고 일조 시간이 적어 채소 생육에 그다지 좋은 환경이 아니기 때문에, 노지 재배보다는 시설 재배가 발달하게 되었다. 이즈음에 네덜란드에서 채소를 중심으로 **수경 재배**가 성행하게 되자 땅에 뿌리를 박고 있지 않은 작물을 사람들이 보고 마치 식물공장 같다고 말했다고 한다. (참고로 [네덜란드](https://korea.investinholland.com/industries/agrifood/)는 미국에 이은 세계 2위의 농업국가이며, 최신기술을 바탕으로 한 농업으로 세계농업시장을 리드하고 있음.)

![식물공장](http://img.asiatoday.co.kr/file/2016y/01m/10d/2016011001000957600050071.jpg)


식물공장이 가시화된 것은 1957년에 덴마크의 크리스텐센 농장에서 새싹채소를 컨베이어 방식으로 생산하면서 부터이다. 이러한 식물공장 방식이 네덜란드, 벨기에, 오스트리아 등으로 확산되었는데, 유럽의 농업 기술을 소개한 일본에서 1970년대에 **식물공장(植物工場)**이라는 용어를 사용하면서 국내에도 같은 용어로 파급된 것으로 보인다. 참고로 일본에는 현재 약 300개소의 식물공장이 상업적으로 운영되고 있는 것으로 알려지며, 미국에서는 [수직농장(vertical farm)](https://en.wikipedia.org/wiki/Vertical_farming) 의 개념이 정립되었다.

식물공장은 한 마디로 온도와 습도를 제어하고 인공 광원으로 농작물을 재배하는 시설농업으로서, 날씨나 계절에 관계없이 농작물을 연중 안정적으로 생산할 수 있다. 따라서 식물공장의 유형은 그 기준에 따라 다양하게 분류할 수 있으며, 개념적으로는 협의와 광의로 구분이 가능하다. 

**-좁은 의미의 식물공장** : 식물을 생산하는 시설 그 자체를 말하며, 주로 수경재배 방식을 기본으로 지상부 제어 범위로는 광, 온‧습도, 탄산가스 등이 해당된다.  

**-넓은 의미의 식물공장** : 조직배양 방법에 의한 클론 증식시스템, 세포배양 방법에 의한 세포 대량증식시스템 등까지 포함하는 개념이다.

이와 같이 식물공장은 농작물의 생육 상태를 과학적으로 관리하여 비료나 농약을 저투입하는 정밀농업(precision agriculture)의 성격을 가지므로, 일반 농산물에 비해 안전성을 확보할 수 있다. 또한 노지에서 재배가 어려운 기능성 농작물을 재배함으로써 고부가가치 농업을 실현할 수 있고, 식량작물의 연중 재배를 통해 생산성을 비약적으로 높임으로써 식량기지로 활용할 수 있다는 점에서 식물공장 방식은 미래 농업의 대안이 될 수 있을 것으로 보인다.

그러나, 장점만 존재하는 것은 아니다. 

**높은 초기 비용** 

LED를 비롯한 값비싼 장비(센서,시스템 구성)로 다른 시설에 비해 훨씬 많은 투자자금이 필요하다. 식물공장 기술이 가장 발달하였고 식품안전에 대한 의식이 높은 일본을 보더라도 식물공장 시공 후 투자자금을 회수하는데 오랜 시간이 필요하였다. 

국내 한 식물공장 업체의 경우 200평 가량의 식물공장 시공에 20억원이 투자되었다. 평당 천 만원의 투자비용이 들어간 셈이다. 

([**※K-Farm 참여 당시 국내 식물공장 운영업체에서 이 부분에 대해 언급했는데, 식물공장을 운영한 지난 10년간 적자생산이었고, 
최근 1-2년 사이에 생산이 안정화되고 유통라인이 자리잡으면서 처음으로 흑자전환 하였다고 말함. 그만큼 초기시설 투자비용을 회수하기가 쉽지 않음**](#))  

지속적인 민간 투자와 정부 보조금이 뒷받침 되지 않아 폐업을 하는 사례도 볼 수 있다. 높은 초기 비용으로 투자 회수 기간이 길기 때문에 투자를 꺼릴 수 밖에 없어 투자 유치가 쉽지 않다.

![미국식물농장폐업사례](http://inews24.cache.iwinv.net/image_gisa/201801/1517281622773_1_121113.jpg)

**생산효율과 유통라인 확보**

식물 1주 당 생산량이 낮은 점은 여전히 식물공장이 떠안고 있는 문제다. 식물공장에서 주로 재배되는 잎채소(상추)는 보통 1주당 100그램의 무게가 나가는데 온실 수경 재배의 200그램 이상, 노지재배 250그램 이상에 비해 많이 부족하다. 심지어 개량된 품종을 노지에서 재배하면 400그램을 넘어가는 것도 흔하다. 

상추 시장 가격은 무게로 책정되기 때문에 1주 당 무게가 낮으면 매출액이 낮을 수밖에 없다. 여기에 기술이 뒷받침되지 않아 재배 과정에서 발생되는 시행착오를 조기에 극복하지 못하면 장기간의 적자로 이어질 수 있다. 

국내 식물공장 기업의 경우 이러한 점을 보완하기 위해, 바질,버터헤드 등 특수작물을 생산하고, 생산품을 가공해 유통라인을 구축하여 적자에 대처하였다.  
 
그러나 이러한 문제점들을 갖고 있음에도 식물공장 산업은 지속적으로 성장한 것으로 전망되고 있다. 

![식물공장전망](http://english.hankyung.com/hkengdata/images/photo/201602/AA.11256112.1.jpg)

농촌경제연구원의 자료에 따르면 세계 식물공장 시장규모는 **2017년 2조 6030에서 2022년 약 2조 4826로** 커질 전망이다.

현재 식물공장의 세계 시장 현황은 어떠한지 살펴보자.

[일본](#)

:이미 지난 1970년대부터 식물공장 개발에 박차를 가했던 일본은 1980년대와 1990년대에 각각 1,2차 식물공장 설립 붐이 일었을 정도다. 

2009년 타업종 기업들의 식물공장 설립 진입 장벽을 낮추는 내용의 개정 농업진흥지역 정비법이 시행되면서 3차 붐이 나타났다. 

2000년대 들어 한국과의 반도체 경쟁에서 뒤진 도시바가 기존에 사용하던 Clean Room을 활용하여 식물공장 산업에 뛰어들었던 것이 대표적인 예이다. 

일본시설원예협회가 조사한 자료에 따르면 일본 내 식물공장은 400여 개 안팎에 달한다. 
이중 절반가량(49.3%)가 2010년 이후부터 재배를 시작했다.

2010년 이후 양적으론 크게 성장했지만 식물공장들의 경영 상황은 좋지 못했다. 전체 식물공장 중 흑자를 내는 곳은 4분의 1에 불과했다. 

42%는 적자에 시달렸으며, 간신히 수지 균형만 맞추고 있는 곳도 33%였다. 

2009년 일본 농림수산성과 경제산업성의 연구 결과에 따르면 10에이커(4만468㎡) 규모의 식물공장을 세울 경우 

일반 비닐하우스에 비해 건축비는 17배가 비쌌고 운영 비용은 47배나 뛰었다. 가격 경쟁력을 갖추기 힘든 구조다.

이러한 구조적인 문제를 넘지 못해 도시바가 식물공장 사업에서 손을 뗐고, 

2015년에는 업계 최대 기업이었던 '미라이社' '산이치팜' 등이 도산하면서 위기를 맞이했다. 

그러나 최근 전자기기 생산업체 히타치와 플랜트 건설업체 닛키, 식품가공업체 카고메 등 주요 대기업이 식물공장에 대한 투자를 늘리면서 

부활 조짐이 보이고 있다. 

이는 기후변화와 농업인구 감소로 인한 농업 생산성 저조라는 문제점을 안고 있는 상황에 지난 30년여간 축적되어온 기술력과 업력을 바탕이 되어 

다시금 식물공장 산업이 재조명 받을 수 있었던 것으로 보인다.

**유럽** 


[네덜란드](#)

사실 네덜란드는 쌀쌀하고, 일조량이 많지 않아 자연조건으로 농업하기에 최적화되어 있지 않다. 
네덜란드는 ‘자연조건이 불리하면 만들면 된다’라는 역발상적인 접근방식으로 농업에 대한 연구가 시작됐고, 그로 인해 농업강국이 되었다. 

농업에 대한 오랜 연구와 개발로 인해 네덜란드는 식물공장의 강자가 됐다. 농촌진흥청이 발표한 보고서 ‘식물 공장의 전망과 정책 과제’에 따르면 네덜란드의 식물공장 재배면적은 1975년에 이미 2008년 세계수준을 앞섰다고한다. 그만큼 기술력이 매우 앞서나가고 있다. 

네덜란드 식물공장 산업의 특징은 다음과 같다. 

**-국가에서 적극적으로 산업육성/선택 집중**

**-불황 시 구제조치보다 산업에 신규기술 개발 및 보급 촉진에 주력**

**-신기술 보급에 관해서는 신규기술을 도입하는 생산자에게 보조금을 지급/세율 컨트롤을 실시**

**-대규모화를 촉진하기 위해서 대규모 공장에 대한 에너지 관련 세율 인하**

**-환경 관련 연구개발에 적극적: 환경 목표의 제정**

**-생산자, 종자메이커, 관련설비 메이커, 금융기관, 대학, 연구기관 등에 의한 협력체제 시스템**

이러한 잘 짜여진 시스템 농업하에 높은 생산성과 소득을 유지하고 있다. (네덜란드는 농업으로만 42억유로의 이익을 창출한다) 

![네덜란드농업클러스터](http://news.kotra.or.kr/images/upload/borawebedit/2015/5/31/180044/0531_!.jpg)


네덜란드는 그린포트(Green Ports)라고 불리는 6개의 클러스터를 중심으로 원예산업이 발달되어 있다. 

이 클러스터에서는 앞서 특징에서 소개했던 바와 같이 기업과 연구기관, 정부가 산·학·연 협업을 이루며 기술혁신을 추진하고 물류를 비롯한 기반시설이 제공된다. 그래서 이곳을 황금 삼각형(golden triangle)이라 부르기도 함. 


**시사점**

기후변화와 농업인구 감소에 대비하기 위한 방안이 필요하기 때문에 ICT를 기반으로 한 농업 분야 연구가 매우 활발하다. 
그러나, 현재 성공적으로 식물공장을 운영하는 국가들의 공통점은 오랜시간 동안 축적한 연구와 기술력이 바탕이 되었다는 것이다. 
또한, 오로지 1차산업으로써의 농업이 아니라 새로운 산업의 형태로써 산학협력등이 이루어지고 제대로 된 유통망이 구축되어야만 확실한 대안으로
자리매김 할 것으로 보인다.
